```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#set directory for line chart
df1=pd.read_excel(r'C:\Users\tonyl\Documents\Parks.xlsx', 'Industry Data')

#set the line chart
plt.plot(df1['Year'],df1['Revenue ($ million)'])
plt.xlabel('Year')
plt.ylabel('Revenue in Million')
plt.title('Annual Revenue from State and National Parks in the US')

plt.show()
```

# Annual Revenue from State and National Parks in the US

The line chart above demonstrates the annual revenue from state and national parks in the U.S.  We can clearly identify several ups and downs. Before we dive into the analysis, we should understand 2 things. First, most parks are nonprofit entities, so government fundings consists a big portion of its revenue. Even though there are traditional for-profit operators in the industry, they make up a small share of the industry. So, the economy could has a big impact on the revenue. Secondly, parks are generally underfunded. Both features can be observed in the line chart. From 2002 to 2008, the parks did not have enough funding as the economy went low and the 2008 recession tanked the revenue to the lowest point. In order to incentify the econmy by rasing the recreational activities, the government increased the funds to parks for 2 years. Once the economy recovered and the revenue picked up, the government decreased the funds to parks and the revenue just freefell again (2010 to 2015). The circle keeps reepeating until 2020 when COVID hit the U.S. and the revenue reached a even lower point, as the tourists stop visiting and recreational activities are not the focus.

```python
# Set Pie Chart1

plt.suptitle('Major Markets', size=16, y=1.12); 

df2=pd.read_excel(r'C:\Users\tonyl\Documents\Parks.xlsx', 'Major Markets')
marketlabel=["Day trips by local residents","Day trips by nonlocal visitors","Overnight trips"]
plt.pie(df2['Revenue%'],labels=marketlabel,radius=1.5,autopct='%0.1f%%',shadow=True, explode=[0,0.1,0.1])

plt.show()
```

# Major Markets
So, for the profitable part of the revenue, what are the major markets for the parks in the U.S.? As shown in the pie chart above, the majority of the market comes from the day trip by local visitors. Following is the day trips by nonlocal visitors. Lastly, is the overnight trips. 

```python
# Set Pie Chart2

plt.suptitle('Revenue Structure', size=16, y=1.12); 

df3=pd.read_excel(r'C:\Users\tonyl\Documents\Parks.xlsx', 'Industry vs Sector Costs')
marketlabel2=["Wages","Purchases","Profit","Depreciation","Marketing","Rent","Utilities","Other"]
quants = []
for index in range(8):
    quants.append(df3['Cost 2%'][index])
plt.pie(quants,labels=marketlabel2,radius=1.5,shadow=True, explode=[0.1,0.1,0,0,0,0,0,0.1])

plt.show()
```

# Revenue Structure
After gaining some understanding of the revenue, we further analyzed the revenue of the parks and tried to understand how the money was spent. As shown in the pie chart above, Wages and the other miscellaneous expense are the 2 leading components of the revenue, followed by the new purchase. These three consist of nearly 45% of the park's revenue. As mentioned above, parks are mostly non-profit organizations, so only a small portion was left as profits. According to the Federal Regulation, these profits must be invested into reinvested into maintenance of their land and buildings next year.

```python
# Set Bar Chart

plt.suptitle('Employment (People) by Year', size=16, y=1); 

df4=pd.read_excel(r'C:\Users\tonyl\Documents\Parks.xlsx', 'Industry Data')
data4_1 = df4['Employment (People)']
data4_2 = df4['Wages ($ million)']
labels = df4['Year']
plt.bar(labels, data4_1)
plt.show()
```

# Employment in Parks
Since wages are such a big part of the total revenue, we draw the bar chart above to understand how many employment numbers in the parks industry. Similar to the revenue line chart, employment has a positive correlation with the revenue. When the revenue is high (parks receive sufficient funds), the employment also goes up. But when the economy was going down, the employment declines. Over the past 5 years, employment declined by 1% due to insufficient funding. Different from the revenue chart, the employment bar chart doesn't show dramatic turns like revenue. Surprisingly, the parks industry can always bring more than 100,000 employment opportunities into the market, with one exception in 2015. Why? Over the year, inadequate funding has made the operators seek different ways to hire people to maintain the parks. many operators have begun relying on part-time and volunteer workers or operating their establishments at reduced hours. These part-time opportunities were also included as employment, which makes the bar chart comparably stable.

```python
# Set Histogram

df5 = pd.read_excel(r'C:\Users\tonyl\Documents\Parks.xlsx', 'Revenue Volatility')
data5 = df5['Annualized Revenue Growth (%)']
plt.hist(data5, bins=30, facecolor="blue", edgecolor="black", alpha=0.7)

#titel
plt.title("Annualized Revenue Growth (%)")
plt.show()
```

# Revenue Growth Analysis
At this point, we have our first conclusion: the parks in the U.S. are not doing so well. The histogram above shows the average annual growth in different industries. As shown above, most industries could reach 2.5 revenue increase in the past 10 years. The parks, on the other hand, is suffering from a negative annual growth with -0.01%. The industry has struggled to maintain consistent revenue growth but it will only be worse in the coming year due to COVID.

```python
# Set Scatter Chart

df6 = pd.read_excel(r'C:\Users\tonyl\Documents\Parks.xlsx', 'Key Ratios')
data_x = df6['Revenue per Employee ($\'000)']
data_y = df6['Wages/Revenue (%)']

#label
plt.xlabel("Revenue per Employee ($\'000)")
plt.ylabel("Wages/Revenue (%)")

plt.scatter(data_x, data_y, alpha=0.6)
plt.show()
```
